# Kalman Filter
A single-dimensional Kalman filter.\
Credit gives to https://github.com/rlabbe/Kalman-and-Bayesian-Filters-in-Python

**A Gaussian data structure**

```python
from collections import namedtuple
gaussian = namedtuple('Gaussian', ['mean', 'var'])
gaussian.__repr__ = lambda s: 'ùí©(Œº={:.3f}, ùúé¬≤={:.3f})'.format(s[0], s[1])
```

```python
g1 = gaussian(3.4, 10.1)
print(g1, g1.var)
```

**Predict function**

```python
def predict(pos, movement):
    return gaussian(pos.mean + movement.mean, pos.var + movement.var)

pos = gaussian(10., .2**2)
move = gaussian(15., .7**2)
predict(pos, move)
```

**Update function**

```python
def gaussian_multiply(g1, g2):
    mean = (g1.var * g2.mean + g2.var * g1.mean) / (g1.var + g2.var)
    variance = (g1.var * g2.var) / (g1.var + g2.var)
    return gaussian(mean, variance)

def update(prior, likelihood):
    posterior = gaussian_multiply(likelihood, prior)
    return posterior

# test the update function
predicted_pos = gaussian(10., .2**2)
measured_pos = gaussian(11., .1**2)
estimated_pos = update(predicted_pos, measured_pos)
estimated_pos
```

**A linear process model**

```python
velocity = 0.1
dt = 0.1
process_var = 1.

process_model = gaussian(velocity*dt, process_var) # displacement to add to x
```

**An example**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t_horizon=10
x = gaussian(10., 1.**2)  # initial position, arbitrary set to show robustness of Kalman filter
sensor_var = 2.

t_list=[]
mean_list=[]
std_list=[]
measurement_list=[]
true_x_list=[]

true_movement = lambda t: velocity*t  # not necessarily the same as the process model

for t in range(int(t_horizon/dt)):
    z = np.random.normal(true_movement(t), np.sqrt(sensor_var))  # generate noisy measurement for true movement
    prior = predict(x, process_model)
    likelihood = gaussian(z, sensor_var)
    x = update(prior, likelihood)
    
    t_list.append(t)
    mean_list.append(x.mean)
    std_list.append(np.sqrt(x.var))
    measurement_list.append(z)
    true_x_list.append(true_movement(t))


def plot_with_fill(x, mean, std, label=None, color=None):
    y_upper=mean+std
    y_lower=mean-std
    if color is not None:
        plt.fill_between(
        x, list(y_lower), list(y_upper), interpolate=True, facecolor=color, linewidth=0.0, alpha=0.3
    )   
    else:
        plt.fill_between(
        x, list(y_lower), list(y_upper), interpolate=True, linewidth=0.0, alpha=0.3
    )     
    plt.plot(x, mean, color=color, label=label) 
    
plt.figure(figsize=(15,8))
plt.scatter(t_list, measurement_list, marker='x', color='red', label='Measurement')
plt.scatter(t_list, mean_list, marker='o', color='blue', label='Kalman Filter Estimation')
plt.plot(t_list, true_x_list,'--', color='grey', label='True Trajectory')
plot_with_fill(t_list, np.array(mean_list), np.array(std_list), label='Kalman Filter')
plt.xlabel('Time')
plt.ylabel('Position')
plt.legend()
plt.grid()
plt.show()
    

 
```

**Hyperparameters**: 

`process_var` and `sensor_var`: both need to be estimated in practice.\
`process_model`: if linear, cannot handle non-linear cases.
